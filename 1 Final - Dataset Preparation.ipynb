# Data for Land Area Neediness

```python
df_area = df_area[['Country', 'Abbreviation', 'Land Area(Km2)']]
df_area.rename(columns = {'Abbreviation': '2_code'}, inplace=True)

df_area.info()
```

## Handling Missing Values

Filling null manually by information provided on internet

```python
null_area = df_area.isnull()

df_area[null_area.any(axis=1)]
```

```python
# filling the missing '2_code' values manually by seeing country code report

country_code_mapping = {
    "Republic of the Congo": "CG",
    "Eswatini": "SZ",
    "Vatican City": "VA",
    "Republic of Ireland": "IE",
    "Namibia": "NA",
    "North Macedonia": "MK",
    "Palestinian National Authority": "PS"
}

df_area['2_code'] = df_area['2_code'].fillna(df_area['Country'].map(country_code_mapping))
```

```python
# filling the only missing values on Land Area (Palestina Country) with the country land area manually
df_area['Land Area(Km2)'].fillna('6020', inplace=True)

# changing the Vatican area from 0 to 1, as actually it have area as big as 0.49km
df_area.loc[df_area['Country'] == 'Vatican City', 'Land Area(Km2)'] = '1'
```

>The Vatican has an area of 0.49 square kilometers, but it's rounded down to 0. It's not possible for it to have zero land area, so I rounded it up. The same goes for rounding when it comes to the population count.

```python
other_zeros = df_area[(df_area['Land Area(Km2)'] == '0')]

if not other_zeros.empty:
    print("There are other rows with a 'Land Area(Km2)' value of 0:")
    print(other_zeros)
else:
    print("There are no other rows with a 'Land Area(Km2)' value of 0.")
```
